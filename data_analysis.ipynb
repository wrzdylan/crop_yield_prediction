# Gathering and Data cleaning

## Pesticides

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams["figure.autolayout"] = True
plt.rcParams['font.size'] = '20'
```

```python
df_pesticides = pd.read_csv('./data/pesticides.csv').pivot_table(index='Year', values='Value', columns='Area')
df_pesticides.head()
```

```python
print( 'How many areas ?', len(df_pesticides.columns) )
print( 'First year ?', df_pesticides.index.min() )
print( 'Last year ?', df_pesticides.index.max() )
```

```python
df_pesticides.describe()
```

### Find and replace nan values

#### How many nan values ?

```python
print(f'There are {df_pesticides.isna().sum().sum()} nan values')
```

#### Find areas with nan values

```python
df_with_nan = df_pesticides.loc[:, df_pesticides.isna().any()]
print(f'There are {len(df_with_nan.columns)} areas with nan values')
df_with_nan.head()
```

```python
df_with_nan.columns
```

#### Fill The former Yugoslav Republic of Macedonia nan values with Yugoslav SFR column

```python
df_pesticides['The former Yugoslav Republic of Macedonia'] = df_pesticides['The former Yugoslav Republic of Macedonia'].fillna(0) + df_pesticides['Yugoslav SFR'].fillna(0)
df_pesticides['The former Yugoslav Republic of Macedonia'].head()
```

#### Fill Sudan nan values with Sudan(former) column

```python
df_pesticides['Sudan'] = df_pesticides['Sudan'].fillna(0) + df_pesticides['Sudan (former)'].fillna(0)
df_pesticides['Sudan'].head()
```

#### Fill Belgium-Luxembourg nan values with Belgium and Luxembourg columns

```python
df_pesticides['Belgium-Luxembourg'] = df_pesticides['Belgium-Luxembourg'].fillna(0) + df_pesticides['Belgium'].fillna(0) + df_pesticides['Luxembourg'].fillna(0)
df_pesticides['Belgium-Luxembourg'].head()
```

### Drop unused areas

```python
df_pesticides = df_pesticides.drop(['Sudan (former)', 'Belgium', 'Luxembourg', 'Yugoslav SFR'], axis=1)
df_pesticides.head()
```

### Fill nan

```python
df_pesticides = df_pesticides.bfill()
df_pesticides.loc[:, df_pesticides.isna().any()].mean()
```

### Data exploration

```python
year, area = np.where(df_pesticides.values == np.nanmax(df_pesticides.values))
```

```python
print( f'Max pesticides ? {np.nanmax(df_pesticides.values)} tonnes used in {df_pesticides.index[year[0]]} by {df_pesticides.columns[area[0]]}' )
```

There are too many areas to plot, then we take the 5% to have the biggest pesticides users.

```python
top_10_pesticides = df_pesticides.loc[:, df_pesticides.mean() > df_pesticides.mean().quantile(.95)]
top_10_pesticides.plot();
```

```python
plt.pie(top_10_pesticides.mean().values, labels=top_10_pesticides.columns, wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
my_circle = plt.Circle( (0,0), 0.6, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
sns.boxplot( x=top_10_pesticides["China, mainland"].values, color='#2ca02c' )
sns.boxplot( x=top_10_pesticides["United States of America"].values, color='#bcbd22' );
```
